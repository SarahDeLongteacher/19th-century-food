```python
#analysing christmas and new years dinners components across time, 1840s to 1940s
#source text https://www.foodtimeline.org/christmasmenu.html#pioneerchristmas

import re
from collections import Counter

def split_into_words(any_chunk_of_text):
    lowercase_text = any_chunk_of_text.lower()
    split_words = re.split("\W+", lowercase_text)
    return split_words

filepath_of_text = "/Users/sarah/Documents/code/holiday_food.txt"

#pulling the most common dishes, by first pulling most used words 
number_of_desired_words = 100

stopwords = ['1', '2', '3', '8', 'a', 'all', 'an', 'and', 'are', 'as', 'assorted', 'at', 'au', 'be', 
             'before', 'book', 'but', 'by', 'christmas', 'cold', 'company', 'cook', 'cookery', 'could', 'course', 
             'day', 'december', 'dinner', 'dinners', 'edition', 'food', 'for', 'from', 'good', 'green', 
             'half', 'hard', 'has', 'have', 'home', 'hot', 'hotel', 'house', 'ice', 'if', 'in', 'is', 'it', 
             'la', 'little', 'made', 'may', 'menu', 'menus', 'more', 'mrs', 'new', 'no', 'not', 'note', 'of', 
             'on', 'one', 'or', 'other', 'our', 'p', 'place', 'put', 's', 'served', 'shell', 'should', 'so', 
             'source', 'table', 'than', 'that', 'the', 'they', 'this', 'three', 'times', 'to', 'two', 'we', 
             'well', 'what', 'when', 'which', 'white', 'will', 'with', 'without', 'would', 'york']

#continue to add stopwords as list output changes then alphabetize manually for my clarity 
stopwords.sort()
print(stopwords)
```

```python
#open and read the file
full_text = open(filepath_of_text, encoding="utf-8").read()

#break down the text and identify most used words and display
all_the_words = split_into_words(full_text)
popular_foods = [word for word in all_the_words if word not in stopwords]
popular_foods_tally = Counter(popular_foods)
most_eaten_food = popular_foods_tally.most_common(number_of_desired_words)

most_eaten_food
```
